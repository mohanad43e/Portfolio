```python
def main(page):
    src = page.content
    soup = BeautifulSoup(src, 'lxml')
    table_head = soup.find('thead')
    table_body = soup.find('tbody').find_all('div', {'class': "L0MOmM"})
    column_header = table_head.contents[0].find_all('div', {'class': "L0MOmM"})
    column_header = [i.text.strip() for i in column_header]
    table_rows = [i.text.strip() for i in table_body]
    table_rows = np.transpose(table_rows)
    table_rows = table_rows.reshape(10,8)
    column_header = np.array(column_header)
    column_header = column_header.reshape(1,8)
    table = np.concatenate((column_header, table_rows), axis = 0)
    table = pd.DataFrame(table)
    table.columns = table.iloc[0]
    table = table.drop([0], axis = 0)
    return table
```

```python
main(page)
```
