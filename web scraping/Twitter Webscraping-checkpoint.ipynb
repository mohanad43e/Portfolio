```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
website = "https://twitter.com/search?q=python&src=typed_query"
path = 'chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get(website)
```

```python
def get_tweet(element):
    user_name = element.find_element(By.XPATH, ".//span[contains(text(), '@')]").text
    text = element.find_element(By.XPATH, ".//div[@lang]").text
    tweet_list = [user_name, text]
    return tweet_list
```

```python
user_name = []
text = []
```

```python
max_scroll = 4
scroll_count = 1
while scroll_count <= max_scroll:
    time.sleep(2)
    driver.execute_script("window.scrollBy(0,document.body.scrollHeight);")
    scroll_count = scroll_count + 1
    tweets = driver.find_elements(By.XPATH, '//article[@role = "article"]')
    time.sleep(2)
    for tweet in tweets:
        tweet_list = get_tweet(tweet)
        user_name.append(tweet_list[0])
        text.append(" ".join(tweet_list[1].split()))
```

```python
tweets_df = pd.DataFrame({"user_name": user_name, "text": text})
```

```python
tweets_df = tweets_df.drop_duplicates().dropna()
tweets_df
```
